[Q10] Complete the function to apply `ratio` column that calculates from `First Class` and `Standard Class` columns on `pivot_table_result`


```python
# [Q10] Complete the function to apply `ratio` column that calculates from `First Class` and `Standard Class` columns on `pivot_table_result`

# function to transform the ratio
def get_class_ratio(row):
    
    # get the first class column
    first_class = row['First Class']
    
    # get the standard class column
    standard_class = row['Standard Class']
    
    # calculate the ratio
    if standard_class != 0:
        ratio = first_class/standard_class
    
    return ratio

pivot_table_result["ratio"] = pivot_table_result.apply(get_class_ratio, axis=1)
    
pivot_table_result.head()
```

[Q11] After complete Q10, What does the `apply` function do?

It call the get_class_ratio function to calculate a ratio in each row of pivot_table_result by using values in First Class and Standard Class column. And append the result to the new column call ratio

[Q12] Create a new column(`short_ratio`) that works the same as Q10 but with `lambda` function


```python
# [Q12] Create a new column(`short_ratio`) that works the same as Q10 but with `lambda` function

pivot_table_result["short_ratio"] = pivot_table_result.apply(lambda row: row['First Class']/row['Standard Class'] if row['Standard Class'] != 0 else None, axis=1)

pivot_table_result.head()
```

[Q13] What is the difference between `apply` and `lambda` function? give 2 examples use case.

apply function is used to apply a function along the axis (rows or columns) of a DataFrame or Series. It can work with any function, including custom functions.
lambda function often used when you want to perform simple operations or pass a function as an argument to another function (like apply).

apply function use case: Applying a custom function across rows (axis=1) or columns (axis=0) to perform a complex operation like summing two columns.

lambda function use case: Using lambda within apply to perform a simple operation, like adding two columns together, without the need to define a separate function.
